```r
install.packages("OECD")
library(tidyverse)
library(httr)
library(jsonlite)
library(dplyr)
library(OECD)
library(ggplot2)
```

```r
df <- get_dataset("NAAG", "CAN+FRA+DEU+ITA+JPN+GBR+USA.DBTEQS12", start_time = 1995, end_time = 2013)
```

```r
print(df)
```

```r
title = "Financial corporations debt to equity ratio \n grouped by G7 member states, from 2004-2013"
df %>%
    ggplot( aes(x=obsTime, y=obsValue, group=LOCATION, color=LOCATION)) +
        geom_line() +
        geom_point() +
        ggtitle(title) +
        theme(plot.title = element_text(size = 20))+
        xlab("year") +
        ylab("debt to quity ratio") +
        theme_classic(base_size = 14) +
        scale_x_discrete(breaks=c(1995, 2000, 2005, 2010, 2013))
```
