Both representations (barplot and jitter) have advantages and disadvantages: <div></div>
<li>The barplot does not provide visual information about the distribution of the data, but is very unambiguous.</li>
<li>The jitterplot visualizes the distribution of the data, but the horizontal shift of the dots can not be interpreted and can thus be confusing to the observer.</li>

<h3> 3) ANOVA </h3>

Grouped by state: differ the mean values of the states significantly?

```r
print("normality test on all groups:")
shapiro.test(df$obsValue)
print("\nnormality test per group:")
locations <- list("ITA", "DEU", "FRA", "JPN", "GBR", "USA", "CAN")
for(l in locations) {
    print(l)
    df_cur <- df[df$LOCATION == l,]
    shapiro.test(df_cur$obsValue)
}
shapiro.test((df[df$LOCATION == "ITA",])$obsValue)
shapiro.test((df[df$LOCATION == "DEU",])$obsValue)
shapiro.test((df[df$LOCATION == "FRA",])$obsValue)
shapiro.test((df[df$LOCATION == "GBR",])$obsValue)
shapiro.test((df[df$LOCATION == "USA",])$obsValue)
shapiro.test((df[df$LOCATION == "CAN",])$obsValue)
shapiro.test((df[df$LOCATION == "JPN",])$obsValue)


```

Caution: p-value not < 0.05 for all groups (USA, CAN, ITA)! This means that the distribution of the data significantly differs from the normal distribution. We continue anyways.
